## Problems

- At the moment I have two problems:
- 1. When I run the example I cannot set the volume of the reactor (`AttributeError: can't set attribute 'volume'`) (see below)
- 2. When I leave out the volume and run a test simulation it takes more than 45 min and is still not finished. I haven't managed to run a successful test simulation yet

````python
from CADETProcess.processModel import ComponentSystem
component_system = ComponentSystem(['A', 'B'])
````

````python
from CADETProcess.processModel import MassActionLaw
reaction_system = MassActionLaw(component_system) 
reaction_system.add_reaction(
    indices=[0,1], 
    coefficients=[-1, 1],
    k_fwd=0.1,
    k_bwd=0
)
````

````python
from CADETProcess.processModel import Cstr
reactor = Cstr(component_system, 'reactor')
reactor.volume = 1e-6
reactor.bulk_reaction_model = reaction_system
reactor.c = [1.0, 0.0]
````

````python
from CADETProcess.processModel import FlowSheet
flow_sheet = FlowSheet(component_system)
flow_sheet.add_unit(reactor)

from CADETProcess.processModel import Process
process = Process(flow_sheet, 'reaction_demo')
process.cycle_time = 3
````

````python
from CADETProcess.simulator import Cadet
simulator = Cadet()
sim_results = simulator.run(process)
_ = sim_results.solution.reactor.outlet.plot()
````
